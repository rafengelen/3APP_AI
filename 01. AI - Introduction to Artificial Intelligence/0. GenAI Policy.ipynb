# Generative AI Policy

Generative AI is a subset of artificial intelligence that can create text, code, images, video, and other content from data provided to it, or retrieved from the internet. In this course, you will have to use generative AI tools for your assignments, such as ChatGPT, Bing AI, Google Bard, code generators, image generators, etc. These tools can help you generate ideas & code, write drafts, create visuals, and more. However, they also pose some challenges and risks that you need to be aware of, and address.

The purpose of this policy is to provide you with some best practices and guidelines on how to use generative AI tools in a secure, responsible, and ethical manner. By following this policy, you will be able to harness the power of generative AI while maintaining academic honesty and integrity.

Each tool usually requires its own style of prompt engineering. **Prompt engineering is the art of crafting effective prompts that elicit high-quality outputs from generative AI tools.** It involves providing clear instructions, relevant information, examples, constraints, feedback loops, etc.

Learning to use these kinds of tools is an emerging skill that requires practice and experimentation. There is no one-size-fits-all approach for using generative AI tools effectively. It comes down to using them yourself and getting a feeling for which tool is a good match for you and your assignments. **Learn by doing, and get better with experience.** Be creative and have fun.

## How to evaluate generative AI outputs

Generative AI tools are not perfect. They can produce outputs that are inaccurate, inconsistent, biased, unethical, or even harmful. Therefore, you need to evaluate the quality and accuracy of the outputs generated by the tools before using them for your assignments.

Here are some tips on how to evaluate generative AI outputs:

- Compare the outputs with other sources. Don't trust anything the tool says without verifying it with other reliable sources. Check for factual errors or inconsistencies with what you know or have learned.
- Look for logical flaws or gaps in reasoning. Generative AI tools can sometimes produce outputs that are nonsensical or illogical. Look for contradictions or fallacies in the arguments or claims made by the tool.
- Consider ethical implications or issues. Generative AI tools can sometimes produce outputs that are offensive or harmful to certain groups or individuals. Consider how the outputs might affect others or reflect your values and principles.
- Use your own judgment and creativity. Generative AI tools are not meant to replace your own thinking or writing skills. They are meant to assist you and inspire you. Use your own judgment and creativity when using the outputs generated by the tools.

## How to cite generative AI outputs

Generative AI outputs are not original works created by you. They are works created by an artificial intelligence tool based on data provided by you or retrieved from the internet. Therefore, you need to cite them properly when using them for your assignments.

Here are some guidelines on how to cite generative AI outputs:

- **Provide a paragraph at the end of your assignment explaining what you used the AI tool for and what prompts you used to get the results**. This is a crucial and necessary part, since each generative tool has some randomness build in. This will help your instructor and peers understand how you used the tool and what inputs you provided. **Failure to do so is in violation of academic honesty policies**.
- Acknowledge the limitations and uncertainties of the outputs. Generative AI outputs are not authoritative or definitive sources of information or knowledge. They are subject to errors, biases, and ethical issues. You need to acknowledge these limitations and uncertainties when using the outputs for your assignments.

## Academic honesty and integrity

Academic honesty and integrity are essential values in any academic setting. They mean that you respect the work of others and take responsibility for your own work. They also mean that you do not engage in any form of plagiarism or misuse of generative AI tools.

Plagiarism is the act of presenting someone else's work as your own, without proper acknowledgment or citation. Misuse of generative AI tools is the act of using them in ways that are inappropriate, unethical, or harmful, such as:

- Using generative AI tools for the practical part of the exam is strictly forbidden, and will result in a zero grade and possible disciplinary action.
- Using generative AI tools to generate entire assignments without any input or effort from you, is cheating and will result in a zero grade and possible disciplinary action. So, make sure you include the end paragraph in each assignment, where your reflect on your use of AI tools, and do not forget to include your prompts used.
- Using generative AI tools to generate outputs that are offensive or harmful to others, is unacceptable and will result in a zero grade and possible disciplinary action.

**You are ultimately responsible for your own work and the use of generative AI tools. You need to use them in a secure, responsible, and ethical manner, following this policy and the general academic honesty policies of this course and this university**.
